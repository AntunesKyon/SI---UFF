```python
from math import sqrt
A = float(input('Digite um valor: '))
B = float(input('Digite um valor: '))
C = float(input('Digite um valor: '))


delta = B**2-4*A*C

if delta >=0:
  print((-B +sqrt(delta))/(2*A))
  print((-B -sqrt(delta))/(2*A))
else:
  print('Não existe solução real')
```

```python
import math

lit_lata = 18 #Totais de Litro na lata
vl_lata = 215 #Valor total da Lata
area_litro =3 #Area de pintura de lara
pintor_hp = 8 #Area total de pintura do Pintor por hora
hora_trabalho_pintor = 60# valor hora do Pintor
rolo_mh = 12 #valor de 2 rolo cada rolo so dura 30min e custa 12

largura = float(input("Digite a largura :"))
comprimento = float(input("Digite a comprimento :"))
altura = float(input("Digite a altura :"))

area = 2*(largura * altura)+2*(largura *comprimento)+(altura*comprimento)

latas = math.ceil(area_litro / lit_lata)

pintura = math.ceil(area/area_litro)
print(f'a area total pinta foi de: {area}m')

total_tempopintura = math.ceil(area/pintor_hp)
print(f'O pintor levara: {total_tempopintura}')

qtd_rolos = math.ceil()(total_tempopintura * 2)
print(f'A quantidade total de rolos usada foi: {qtd_rolos}')

qtd_rolos = math.ceil(total_tempopintura * 2)
print(f'A quantidade totais de rolos necessario e {qtd_rolos}')

custo_tinta = math.ceil(latas * vl_lata)
print(f'O custo total de tinta foi {custo_tinta}')

custo_rolos = math.ceil(qtd_rolos * rolo_mh)
print(f'O total dos rolos foi {custo_rolos}')

custo_mao_obra = math.ceil(total_tempopintura * hora_trabalho_pintor)
print(f'O custo total de mao obra foi {custo_mao_obra}')

custo_total = math.ceil(custo_tinta + custo_rolos + custo_mao_obra)
print(f'O custo total da obra foi de {custo_total}')
```
